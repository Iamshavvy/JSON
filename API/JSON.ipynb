```python
import requests
import json

url = "http://api.codingninjas.com/api/v3/courses"
response = requests.get(url)

if response.status_code == 200:
    data = json.loads(response.text)
    for course in data['data']['courses']:
        if course['id'] == 19:
            print(course['title'])
else:
    print("Error occurred while fetching the data")
```

```python
import requests

url = "https://api.openaq.org/v2/cities"
params = {
    "country": "AU",
}
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    cities = data["results"]
    count = len(cities)
    print(count)
else:
    print("Error occurred while fetching the data")
```

```python
url='https://api.openaq.org/v1/cities'
params ={
    "country" : "BE",
    "limit" : 5 ,
    "order_by": "city"
}
response=requests.get(url,params=params)

data=response.json()
cities=data['results']
for value in cities:
    print(value['city'])
```
